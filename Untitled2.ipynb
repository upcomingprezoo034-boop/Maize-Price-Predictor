```python
from flask import Flask, request, jsonify
import joblib

app = Flask(__name__)

model = joblib.load("price_prediction_model.pkl")

@app.route('/predict', methods=['POST'])
def predict():

    data = request.json

    rainfall = data['rainfall']
    season = data['season']

    prediction = model.predict([[rainfall, season]])

    return jsonify({
        "predicted_price": float(prediction[0])
    })

app.run()
```

```python
%%writefile app.py
import streamlit as st
import joblib

# Load trained AI model
model = joblib.load("price_prediction_model.pkl")

st.title("Climate-Smart Crop Price Prediction Tool")

st.write("Predict future market prices for dryland farmers")

# User Inputs
rainfall = st.number_input("Enter Rainfall (mm)", min_value=0.0)

season = st.selectbox(
    "Select Season",
    ("Planting", "Growing", "Harvest")
)

# Convert season to numeric
season_map = {
    "Planting":0,
    "Growing":1,
    "Harvest":2
}

season_value = season_map[season]

# Prediction button
if st.button("Predict Price"):

    prediction = model.predict([[rainfall, season_value]])

    st.success(f"Predicted Market Price: {prediction[0]:.2f} KES/kg")

    if prediction[0] > 120:
        st.write("Recommendation: Sell produce now")
    else:
        st.write("Recommendation: Store produce and wait")
```
